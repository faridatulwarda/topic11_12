Nilai `std` Region Tangerang dan Jakarta Pusat bagian Latitude  nilai max nya  melewati  dari Q3.

```python
df_jp = df_update[df_update['Region']=='Jakarta Pusat']
Q1 = df_jp['Latitude'].quantile(0.25)
Q3 = df_jp['Latitude'].quantile(0.75)
IQR = Q3 - Q1

# Cari data yang diluar outlier 
outlier_jp = df_jp.query("Latitude < (@Q1 - 1.5 * @IQR) or Latitude > (@Q3 + 1.5 * @IQR)")
outlier_jp   
```
